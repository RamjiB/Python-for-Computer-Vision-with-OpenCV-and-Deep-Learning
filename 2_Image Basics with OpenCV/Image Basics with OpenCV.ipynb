### Class 3- Drawing on Images

```python
# create a blank image
blank_img = np.zeros(shape=(512,512,3),dtype=np.int16)
```

```python
plt.imshow(blank_img)
```

```python
blank_img.shape
```

```python
#draw rectangle on the blank image
#pt1 - top left corner
#pt2 - bottom right corner
cv2.rectangle(blank_img,pt1 =(200,200),pt2=(300,300),
              thickness=10,color=(255,0,0))
```

```python
plt.imshow(blank_img)
```

```python
## draw circles
cv2.circle(blank_img,center=(256,256),
           radius=50,color=(0,255,0),thickness=5)
```

```python
plt.imshow(blank_img)
```

```python
## to draw filled in circles are images give thickness as negative 1(-1)
cv2.circle(blank_img,center=(400,400),radius=50,color=(0,255,0),thickness=-1)
plt.imshow(blank_img)
```

```python
## to draw line
cv2.line(blank_img,pt1=(0,0),pt2=(512,512),color=(0,0,255),thickness=3)
plt.imshow(blank_img)
```

```python
## add text to image
font = cv2.FONT_HERSHEY_SIMPLEX
cv2.putText(blank_img,text='Hey!!!',org=(10,400),fontFace = font,
            fontScale=4,color=(255,255,255),thickness = 3,\
            lineType=cv2.LINE_AA)
plt.imshow(blank_img)
```

```python
## custom polygons
new_bl_img = np.zeros(shape=(512,512,3),dtype=np.int32)
plt.imshow(new_bl_img)
```

```python
vertices = np.array([[100,300],[200,200],[400,100],[300,400]],dtype=np.int32)
```

```python
vertices
```

```python
# cv2 needs a 3 dimension whereas our vertices are in 2 D
pts = vertices.reshape((-1,1,2))
```

```python
pts.shape
```

```python
cv2.polylines(new_bl_img,[pts],isClosed= True,
             color=(255,255,255),thickness=3)
plt.imshow(new_bl_img)
```

### Class 6

```python
import numpy as np
import cv2

# variables

drawing = False
ix,iy = -1,-1

# Function
def draw_rect(event,x,y,flags,param):
    global ix,iy,drawing
    
    if event == cv2.EVENT_LBUTTONDOWN:
        drawing = True
        ix,iy = x,y
    elif event == cv2.EVENT_MOUSEMOVE:
        if drawing == True:
            cv2.rectangle(img,(ix,iy),(x,y),(0,255,0),-1)
    elif event == cv2.EVENT_LBUTTONUP:
        drawing = False
        cv2.rectangle(img,(ix,iy),(x,y),(0,255,0),-1)

#section for shwoing the image
img = np.zeros((512,512,3),dtype=np.int8)

cv2.namedWindow(winname = 'my_drawing')
cv2.setMouseCallback('my_drawing',draw_rect)

while True:
    
    cv2.imshow('my_drawing',img)
    if cv2.waitKey(20) & 0xFF == 27:
        break
cv2.destroyAllWindows()
```
